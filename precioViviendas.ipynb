```python
#paso 1
!pip install tensorflowjs

#paso 2
!pip install tensorflow==2.16.0rc0

#paso 3
import os
os.environ["TF_USE_LEGACY_KERAS"] = "1"

#paso 4
!mkdir precio_viviendas

#paso 5
import tensorflow as tf
import numpy as np

# datos de entrada
# [tamaño, número de habitaciones, antigüedad]
entradas = np.array([[50, 1, 10], [80, 2, 15], [120, 3, 20], [150, 4, 5], [200, 5, 8]], dtype=float)
precios = np.array([800000, 1500000, 2100000, 2500000, 3500000], dtype=float)  # precios

# definir modelo
modelo = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=[3]),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dense(units=1)
])

# compilacion
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.01),
    loss='mean_squared_error'
)

# modelo entrenado
print("Comenzando entrenamiento...")
historial = modelo.fit(entradas, precios, epochs=300, verbose=False)
print("Modelo entrenado!")

# guardar el modelo en formato HDF5
modelo.save('precio_viviendas.h5')

# Convertir a TensorFlow.js
!tensorflowjs_converter --input_format=keras precio_viviendas.h5 precio_viviendas
```
